#벡터와 array


*목적: 벡터는 주로 수학적, 물리학적 개념을 표현하는 데 사용되는 반면, 배열은 프로그래밍에서 데이터를 효율적으로 관리하기 위해 사용됩니다.   
*추상화 수준: 벡터는 방향과 크기라는 추상적인 속성을 가진 반면, 배열은 구체적인 데이터 요소의 집합으로 구성됩니다.   
*연산: 벡터 연산은 수학적 성질에 기초하여 정의되며, 배열 연산은 주로 데이터 처리와 관련된 작업에 초점을 맞춥니다.

##벡터의 연산

덧셈, 뺄셈, 스칼라곱 scalar product, 성분곱 elementwise multiplication, 내적 inner product

```python
#vector > 프로그래밍에서는 array로 사용한다

v = np.array([1,3,5])
w = np.array([2,4,6])

print('v + w =',  v + w) # 덧셈

```

```python
#vector
v = np.array([1,3,5])
w = np.array([2,4,6])

print('v - w =', v - w) # 뺄셈
```

#행렬의 연산

덧셈, 뺄셈, 스칼라곱 scalar product, 성분곱 elementwise multiplication,

행렬곱 matrix multiplication, 전치 transpose, 단위행렬 identity matrix, 역행렬 inverse matrix

```python
# 행렬 연산 1
X = np.array([[3, 4, 2], [1, 3, 7]])
Y = np.array([[1, 0, 1], [2, 1, 1]])


print('X + Y =\n', X + Y, '\n') # 덧셈


print('X - Y =\n', X - Y) # 뺄셈
```

```python
# 행렬 연산 2

X = np.array([[3, 4, 2], [1, 3, 7]])
Y = np.array([[1, 0, 1], [2, 1, 1]])


print(3 * X) # 곱셈 : 스칼라곱

```

```python
print(np.multiply(X, Y)) # 곱셈 : 성분곱
```

```python
# 행렬 연산 3
#행과 열의 곲 > 2행과 2열의 곱 > 2x2

X = np.array([[1, 0, 3], [2, 1, 4]])
Y = np.array([[2, 3], [1, 2], [0, 1] ])

#1row*1col:1*2=2, 1row*2col:3+3=6, 2row&1col: 4+1=5, 2row*2col: 6+2+4=12
print(X.dot(Y)) # 행렬곱 (2x3)x(3x2)=(2x2)
```

```python
# 전치 (Transpose)
#1행3열 > 3행1열
x1 = np.array([[3], [4], [5]])
#2행3열 > 3행2열
x2 = np.array([
    [1,0,3],
    [2,1,4]
])

print('x1.T : ',x1.T)
print('x2.T : ',x2.T)
```
